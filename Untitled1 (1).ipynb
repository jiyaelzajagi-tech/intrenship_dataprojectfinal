```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Gold-Silver-GeopoliticalRisk_HistoricalData(1).csv')

```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df = df.sort_values('DATE')

plt.figure(figsize=(12, 6))
plt.plot(df['DATE'], df['GOLD_PRICE'], label='Gold Price')
plt.title('Historical Gold Price Trend')
plt.xlabel('Year')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

Observation: The line plot shows clear volatility in gold prices. Future analysis could involve correlating these spikes with specific geopolitical events identified in the 'EVENT' column of the dataset.
